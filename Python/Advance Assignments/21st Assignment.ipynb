1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all ?

    A probability distribution is a function that describes the likelihood of different outcomes occurring in an experiment.It describes how the random variable are distributed. 

    While the individual outcomes may be random, the distribution itself provides information about the likelihood of those outcomes. For example, in a fair six-sided die roll, each outcome has an equal probability of 1/6. So, while we can't predict the exact outcome of a single roll, but we can predict the overall distribution of outcomes over many rolls.

2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?

    True random numbers are generated from physical processes, while pseudo-random numbers are generated by algorithms and thus are deterministic. The latter are considered "good enough" because they can exhibit a uniform distribution and are efficiently produced, making them suitable for various practical applications.

3. What are the two main factors that influence the behaviour of a "normal" probability distribution ?

    The two main factors that influence the behavior of a "normal" probability distribution are the mean (μ) and the standard deviation (σ).

4. Provide a real-life example of a normal distribution ?

 - A real-life example of a normal distribution is the distribution of heights/weights in a population. 

5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows ?
 
    In the short term, a probability distribution can be unpredictable with a small number of trials due to randomness. However, as the number of trials increases, the distribution tends to stabilize, outcomes become more predictable, and converge towards expected probabilities.

9. What are the three logarithmic functions that Python supports ?

    The Three Logarithmic Functions that Python supports are:

    - log2(x): Returns the logarithmic value of x to base 2.

    - log10(x): Returns the logarithmic value of x to base 10.
    
    - log(x, base): Returns the logarithm of x to the given base. If no base is specified, returns the natural logarithm (base e) of x
